```python
df=data
```

```python
# Step 2: Display summary statistics for each column
df.describe()
```

```python
# Calculate measures of dispersion for each quantitative variable
mad = df[["Age", "Annual Income (k$)"]].apply(np.abs).mean()
variance = df[["Age", "Annual Income (k$)"]].var()
std_dev = df[["Age", "Annual Income (k$)"]].std()
range = df[["Age", "Annual Income (k$)"]].max() - df[["Age", "Annual Income (k$)"]].min()
quartiles = df[["Age", "Annual Income (k$)"]].quantile([0.25, 0.5, 0.75])
skewness = df[["Age", "Annual Income (k$)"]].skew()
```

```python
# Calculate summary statistics for income by age group
grouped_df = df.groupby("Age")["Annual Income (k$)"].agg(["min", "max", "mean", "median"])
```

```python
# Print the results
print("\nMeasures of Dispersion:")
print(f"Mean Absolute Deviation:\n{mad}")
print(f"\nVariance:\n{variance}")
print(f"\nStandard Deviation:\n{std_dev}")
print(f"\nRange:\n{range}")
print(f"\nQuartiles:\n{quartiles}")
print(f"\nSkewness:\n{skewness}")
```

```python
# Create new age categories
df["Age Category"] = pd.cut(df["Age"], bins=[18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70], labels=["18-25", "25-30", "30-35", "35-40", "40-45", "45-50", "50-55", "55-60", "60-65", "65-70"])
```

```python
# Group data by age category and calculate summary statistics for income
grouped_df = df.groupby("Age Category")["Annual Income (k$)"].agg(["min", "max"])
```

```python
# Print the results
print("\nSummary Statistics for Income by Age Category:")
print(grouped_df.to_string())
```
