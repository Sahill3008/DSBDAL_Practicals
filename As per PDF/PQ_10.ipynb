2. Detect the outlier using BoxPlot.
3. Handle the outlier using Quantile based flooring and capping (Hint: the outlier is capped
at a certain value above the 90th percentile value or floored at a factor below the 10th
percentile value)


```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the demo dataset into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/DSBDAL_Exam/demo1.csv')
```

```python
# Display the first few rows of the DataFrame to check the data
df.head()
```

```python
# Create a boxplot to visualize the distribution of data and detect outliers
sns.boxplot(data=df)
plt.title('Boxplot of the Demo Dataset')
plt.show()
```

```python
# Define the quantiles for flooring and capping
floor_quantile = 0.1  # Floor at the 10th percentile
cap_quantile = 0.9    # Cap at the 90th percentile
```

```python
# Calculate the flooring and capping values
floor_value = df.quantile(floor_quantile)
cap_value = df.quantile(cap_quantile)
```

```python
# Apply flooring and capping to handle outliers
df = df.clip(lower=floor_value, upper=cap_value, axis=1)
```

```python
# Display the first few rows of the DataFrame after handling outliers
df.head()
```

```python
# Create a boxplot to visualize the distribution of data after handling outliers
sns.boxplot(data=df)
plt.title('Boxplot of the Demo Dataset After Handling Outliers')
plt.show()
```
