Consider the given dataset Students Performance Test1
1. Check that is there any missing values in dataframe as a whole
2. is there any missing values across each column
3. count of missing values across each column
4. count row wise missing value
5. count of missing values of a gender column.
6. groupby count of missing values of a column , consider column gender and score
7. replace the missing value of score column with average value of the column


```python
import pandas as pd
import numpy as np
```

```python
db=pd.read_csv("/content/drive/MyDrive/DSBDAL_Exam/StudentsPerformanceTest1.csv")
```

```python
db
```

```python
# Check if there are any missing values in the dataframe as a whole
print("Any missing values in the dataframe as a whole:", db.isnull().values.any())
```

```python
db.isnull()
```

```python
# Check if there are any missing values across each column
print("\nMissing values across each column:")
print(db.isnull().sum())
```

```python
# Count of missing values across each column
print("\nCount of missing values across each column:")
print(db.isnull().sum())
```

```python
# Count row-wise missing values
print("\nCount of missing values row-wise:")
print(db.isnull().sum(axis=1))
```

```python
# Count of missing values in the 'gender' column
print("\nCount of missing values in the 'gender' column:", db['gender'].isnull().sum())
```

```python
# Groupby count of missing values of a column, considering columns 'gender' and 'score'
print("\nGroupby count of missing values of 'gender', 'Placement Score','writing score','reading score' and 'math score' columns:")
print(db.groupby(['gender', 'Placement Score','writing score','reading score','math score']).size().unstack(fill_value=0))
```

```python
db.replace('Na','NaN', inplace=True)
```

```python
# Calculate average score excluding NaN values
average_score = db['writing score'].mean(skipna=True)
print("Avg Writing Score",average_score)
db['writing score'].fillna(average_score, inplace=True)
print("\n",db)
```

```python
# Calculate average score excluding NaN values
average_score = db['writing score'].mean(skipna=True)
print("Avg Placement Score",average_score)
db['Placement Score'].fillna(average_score, inplace=True)
print("\n",db)
```

```python
db['math score'] = db['math score'].astype(float).round(1)
db['reading score'] = db['reading score'].astype(float).round(1)
```

```python
# Calculate average score excluding NaN values
average_score = db['math score'].mean(skipna=True)
print("Avg Placement Score",average_score)
db['math score'].fillna(average_score, inplace=True)
print("\n",db)
```
