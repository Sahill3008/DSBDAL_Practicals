4. Replace the missing value of Math Score with Mean Value
5. Replace the missing value of Reading Score with standard deviation
6. Replace the missing value of place with common value "Nashik

```python
series = pd.isnull(db["math score"])
```

```python
# Replace null values with NaN
db.fillna(np.nan, inplace=True)
db['math score'] = db['math score'].replace('na', np.nan)
db['reading score'] = db['reading score'].replace('Na', np.nan)
db['Region'] = db['Region'].replace('Na', np.nan)
db['Region'] = db['Region'].replace('na', np.nan)
```

```python
db
```

```python
db['math score'] = db['math score'].astype(float).round(1)
db['reading score'] = db['reading score'].astype(float).round(1)
```

```python
#Replace missing values in Math Score with the mean
db['math score'].fillna(db['math score'].mean(), inplace=True)
db['reading score'].fillna(db['reading score'].mean(), inplace=True)
```

```python
db
```

```python
#Replacing null value with Nashik in Region coloum
db.replace(to_replace = np.nan, value = "Nashik")
```

```python
# Replace missing value of reading score with standard deviation
db['reading score'].fillna(db['reading score'].std(), inplace=True)
```
