1. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the
above statistics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset("titanic")
```

```python
dataset
```

```python
sns.boxplot(x= 'sex',y='age' ,data=dataset)
```

```python
sns.boxplot(x= 'sex',y='age' ,data=dataset,hue="survived")
```

```python
# Plot a box plot for distribution of age with respect to each gender and survival status
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset, x='sex', y='age', hue='survived')
plt.title('Distribution of Age with Respect to Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
