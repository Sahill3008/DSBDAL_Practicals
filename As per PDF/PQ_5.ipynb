1. Locate open Iris dataset from the URL csv_url =
'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
2. Perform the label encoding , by considering Species as target variable.

```python
import pandas as pd
import numpy as np
```

```python
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
data=pd.read_csv(csv_url)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_names =['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']
```

```python
db = pd.read_csv(csv_url, names = col_names)
```

```python
db
```

```python
db['Species'] = le.fit_transform(db['Species'])
```

```python
print("data contains the species setosa, versicolor, and virginica, the encoded values will be 0, 1, and 2, respectively")
db
```
