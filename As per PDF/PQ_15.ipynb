```python
sns.histplot(dataset['fare'], kde=False, bins=10)
```

```python
# Plot histogram for distribution of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(dataset['fare'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
